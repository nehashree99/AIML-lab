```python
year=range(2013,2019) # Changed 2018 to 2019 to generate a list of 6 numbers

apples=[0.895,0.91,0.919,0.926,0.929,0.931]
oranges=[0.962,0.941,0.93,0.923,0.918,0.908]
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.legend(['Apples','Oranges'])
plt.title('Crop yield in kanto')
```

```python
plt.plot(year,apples,marker='o')
plt.plot(year,oranges,marker='x')
plt.ylabel('yeild')
plt.xlabel('years')
plt.title('Crop yield in kanto')
plt.legend(['Apples','Oranges'])
```

```python
#Styling marker
plt.plot(year,apples,marker='s',c='b',ls='-',lw=2,ms=8,mew=2,mec='orange')
plt.plot(year,oranges,marker='o',c='r',ls='--',lw=2,ms=8,alpha=0.5)
plt.ylabel('yeild')
plt.xlabel('years')
plt.title('Crop yield in kanto')
plt.legend(['Apples','Oranges'])
```

```python
plt.plot(year,apples,'s-b',label='Apples')
plt.plot(year,oranges,'o--r',label='Oranges')
plt.ylabel('yeild')
plt.xlabel('years')
plt.title('Crop yield in kanto')
plt.legend()
```

```python
plt.plot(year,oranges,'or')
plt.plot(year,apples,'sb')
plt.ylabel('yeild')
plt.xlabel('years')
plt.title('Crop yield in kanto')
plt.legend(['Apples','Oranges'])
```

```python
plt.figure(figsize=(5,5))
plt.plot(year,oranges,'or')
plt.plot(year,apples,'sb')
plt.ylabel('yeild')
plt.xlabel('years')
plt.title('Crop yield in kanto')
plt.legend(['Apples','Oranges'])
```

```python
import matplotlib.pyplot as plt
x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
```

```python
plt.bar(x,y)
plt.title("Bar Chart")
plt.legend(["Bar"])
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
#scatter plot
import matplotlib.pyplot as plt
x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.legend(["Scatter"])
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
e=(0.1,0.1,0,0)
plt.pie(x,explode=e)
plt.title("Pie Chart")
plt.legend(["Pie"])
plt.show()
```
